# Data Visualisation with Python and Matplotlib

### Import key libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Line Graph

```python
plt.plot([1,2,3,4,5], [4,5,6,7,3])
plt.show()
```

### Labels, titles, legend and window buttons

```python
x = [1,2,3,4,5]
y = [2,6,4,8,1]
y2 = [1,3,5,7,9]

plt.plot(x,y, label='First Graph')
plt.plot(x,y2, label='Second Graph')
plt.xlabel('Plot Number')
plt.ylabel('Random #')
plt.title('Epic Graph Tutorial for data viz in Python with Matplotlib. \nTutorial showing labels and titles')
plt.legend()
plt.show()
```

### Bar Chart

```python
x = [2,4,6,8,10]
x2 = [1,3,5,7,9]
y = [4,7,4,7,3]
y2 = [5,8,3,4,2]

plt.bar(x,y,label='One', color='green')
plt.bar(x2, y2, label='Two', color='blue')

plt.xlabel('Bar Number')
plt.ylabel('Bar Height')
plt.legend()
plt.show()
```

### Histogram

```python
test_score = [56,78,90,91,100,85,76,78,99,98,91,92,90,100,100,99,94]

x = [x for x in range(len(test_score))]

# plt.bar(x, test_score)
plt.hist(test_score, bins=list(np.arange(10,110,10)), histtype='bar', cumulative=False, rwidth=0.8, label='Test Score')
plt.legend()
plt.show()
```

### Scatter Plot

```python
test_score = [56,78,90,91,100,85,76,78,99,98,91,92,90,100,100,99,94]
time_spent = [x*60/np.sum(test_score) for x in test_score]

plt.xlabel('Time Spent % of hours')
plt.ylabel('Test Score')
plt.title('Test Score  vs Time Spent \nAnalysis')
plt.scatter(time_spent, test_score)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [2,3,2,4,2]
y2 = [8,8,6,7,6]

plt.scatter(x, y1, marker='o', color='g')
plt.scatter(x, y2, marker='v', color='m')

plt.show()
```
